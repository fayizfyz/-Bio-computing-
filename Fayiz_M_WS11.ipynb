1. Parse the uniprotSnip.fasta to extract information and store in dictionary. The information to store is:
**UniqueIdentifier, EntryName, OrganismName, GeneName, ProteinExistence, ProteinSequence**


```python
import re

def parse_data(file):
    protein_data = {}
    with open(file, "r") as f:
        content = f.read()

        entries = re.findall(
            r">\w+\|(\w+)\|(\w+)\s(.*?)\sOS=(.*?)\sOX=(\d+)\s(?:GN=(\w+))?\sPE=(\d+)\sSV=(\d+)\n((?:[\s\S]*?))(?=\n>|$)",
            content
        )

        for entry in entries:
            unique_identifier = entry[0]
            protein_data[unique_identifier] = {
                "EntryName": entry[1],
                "ProteinName": entry[2],
                "OrganismName": entry[3],
                "OrganismIdentifier": entry[4],
                "GeneName": entry[5] if entry[5] else None,
                "ProteinExistence": entry[6],
                "SequenceVersion": entry[7],
                "ProteinSequence": entry[8].replace("\n", "")
            }

    return protein_data

file_path = "/content/uniprotSnip.fasta"
protein_data = parse_data(file_path)


print(protein_data)
```

```python
import re
```

#Regular expression


```python
#ID match
with open('/content/uniprotSnip.txt',"r") as f:
  data=f.read()
  pattern= re.compile(r'(ID)\s\s\s([A-Z]\w*)')
  ID_match=pattern.finditer(data)
  #print(matches)
  for mat in ID_match:
    print(mat.group(2))

```

```python
#AC match
with open('/content/uniprotSnip.txt',"r") as f:
  data=f.read()
  pattern= re.compile(r'(AC)\s\s\s([A-Z0-9]\w*)\;')
  AC_match=pattern.finditer(data)
  #print(matches)
  for mat in AC_match:
    print(mat.group(2))
```

```python
#DR match
with open('/content/uniprotSnip.txt', "r") as f:
    data = f.read()
    patterns = re.findall(r'DR\s+(.*?);\s([^\n]*)', data)
    for pat in patterns:
        print(pat[0], pat[1])
```

```python
#SQ match (AA and MW)
with open('/content/uniprotSnip.txt', "r") as f:
    data = f.read()
    patterns = re.findall(r'SQ\s+SEQUENCE\s+(\d+)\s+AA;\s+(\d+)\s+MW;', data)
    for pat in patterns:
        AA=pat[0]
        MW=pat[1]
        print("AA",AA)
        print("MW",MW)

```
